Random Forest from sklearn to fill all NaN in "LocationAccuracy". Feeding 'LocationAccuracy', 'latitude', 'DepthInMeters'. No 'longitude' because I found 'longitude' might cause very large range scale.

```python
from sklearn.ensemble import RandomForestRegressor

def set_missing_LocationAccuracy(df):
    location_df = df[['LocationAccuracy','latitude','DepthInMeters']]
    known_location = location_df[location_df.LocationAccuracy.notnull()].as_matrix()
    unknown_location = location_df[location_df.LocationAccuracy.isnull()].as_matrix()
    y = known_location[:, 0]
    X = known_location[:,1:]
    rfr = RandomForestRegressor(random_state=0, n_estimators=2000, n_jobs=-1)
    rfr.fit(X, y)
    predictedLocation = rfr.predict(unknown_location[:, 1::])
    df.loc[ (df.LocationAccuracy.isnull()), 'LocationAccuracy' ] = predictedLocation 
    return df,rfr

ata_train, rfr = set_missing_LocationAccuracy(data_train_rfr)
```
